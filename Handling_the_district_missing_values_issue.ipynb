# Documentation

The district data had some issue importing.

## **Issue** :
Some cells in the district CSV contain a question mark ? instead of a number. The bank used ? to represent missing values, like how we would use NULL. PostgreSQL does not understand ? as a number so it throws that error.
We are going to fix this using Python. I uploaded the dataset and intend to replace every? with nothing (empty), and save a clean version. Then we import the clean version.

```python
import numpy as np
import pandas as pd
```

```python
# Read the district CSV
df = pd.read_csv('/content/district.csv', sep=';', na_values=['?'])

# Load the head of the dataset
print(df.head())
```

```python
# Check the shape of the data
print(df.shape)
```

```python
# Check for missing values in the dataset
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
# Fill missing values with the column average
df['A12'] = df['A12'].fillna(df['A12'].mean())
df['A15'] = df['A15'].fillna(df['A15'].mean())

```

```python
# Verify no more missing values
print("Missing values after cleaning:")
print(df.isnull().sum())
```

```python
# Save the clean file
df.to_csv('/content/district_clean.csv', sep=';', index=False)
print("\nClean file saved successfully")
```
